```python
figsize = (10, 8)
fig = plt.figure(figsize=figsize)
ax = fig.subplots(1, 1)
title = "title"
ylabel = "ylabel"
labelfontsize = 8

# Cat. A

ticklabels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
height = [15, 30, 45, 10]
x = np.arange(len(ticklabels))    # the x locations for the groups
width = 0.13         # the width of the bars

labels = ["Cat. A"]
bar(height,
    x=x,
    ticklabels=ticklabels,
    color="blue",
    ylabel=ylabel,
    width=width,
    labelfontsize=labelfontsize,
    title=title,
    ax=ax)


# Cat. B
labels = labels +  ["Cat. B"]
height = [8, 23, 11, 15]
bar(height,
    x=x+width,
    ticklabels=ticklabels,
    color="red",
    ylabel=ylabel,
    width=width,
    labelfontsize=labelfontsize,
    title=title,
    ax=ax)


# Cat. C
labels = labels + ["Cat. C"]
height = [8, 23, 11, 15]
bar(height,
    x=x+2*width,
    ticklabels=ticklabels,
    color="green",
    ylabel=ylabel,
    width=width,
    labelfontsize=labelfontsize,
    title=title,
    ax=ax)



# Cat. D
labels = labels + ["Cat. D"]
height = [8, 23, 11, 15]
bar(height,
    x=x+3*width,
    ticklabels=ticklabels,
    color="brown",
    ylabel=ylabel,
    width=width,
    labelfontsize=labelfontsize,
    title=title,
    ax=ax)

# Cat. 5
labels = labels + ["Cat. E"]
height = [8, 23, 11, 15]
bar(height,
    x=x+4*width,
    ticklabels=ticklabels,
    color="grey",
    ylabel=ylabel,
    width=width,
    labelfontsize=labelfontsize,
    title=title,
    ax=ax)


ax.legend(labels)
_ = ax.set_xticks(x + 2*width)
_ = ax.set_xticklabels(ticklabels)
```
